```
fp = open('english_python_data.txt', 'r' )
lines = fp.readlines()
len(lines)
```

```
print(lines[4:12])
```

```
#Removing empty lines
lines = [ l for l in lines if len(l)>1]
len(lines)
```

```
print(lines[4:12])
```

```
data = [['text', 'sourcecode']]
textlines = []
k=0
# Captured the line numbers of the text
for i in range(len(lines)):
  
  if lines[i].startswith('#') and len(lines[i])>25 and i!=(k+1):
    textlines.append(i)
    k=i
print(textlines)
```

```
#Writing question and code to a list and appending them to the list of lists.
for i in range(len(textlines)):
  text = lines[textlines[i]]
  text = text.lstrip('#').strip()
  if (i < len(textlines) - 1):
    code = ''.join(lines[textlines[i]+1 : textlines[i+1]])
  else:
    code = ''.join(lines[textlines[i]+1 : ])
  if code != "":
    data.append([text,code])
  
```

```
import csv

with open('TexttoPython.csv', 'w') as f:
  write = csv.writer(f, delimiter=',')
  write.writerows(data)
```
